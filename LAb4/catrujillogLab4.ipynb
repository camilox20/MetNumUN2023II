## SymPy’s polynomials
- https://docs.sympy.org/latest/modules/polys/basics.html#polynomials

$$ P_n(x) =  a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0$$

$$P_n(x) = 0 $$

$$a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0 = 0$$

$$(x-r_1) Q_{n-1}(x) = P_n(x)$$
$$(x-r_1) (x - r_2) Q_{n-2}(x) = P_n(x)$$
$$ \vdots $$

- x^10 - (341 x^8)/16 + (5797 x^6)/64 - (5797 x^4)/64 + (341 x^2)/16 - 1

```
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**10-(341*x**8)/16+(5797*x**6)/64-(5797*x**4)/64 +(341*x**2)/16-1

p, r = div(p,  x-1)

print(p)
print(r)

p, r = div(p,  x-2)

print(p)
print(r)

p, r = div(p,  x-4)

print(p)
print(r)

p, r = div(p,  x+1)

print(p)
print(r)

```

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = 2*x1 + 3*I*x2,y = 4*y1 + 5*I*y2, z = z1 + 6*I*z2, x * y * z

image.png

```
from sympy import symbols, I, N, E, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)
x = 2*x1 + 3*I*x2
y = 4*y1 + 5*I*y2
z = z1 + 6*I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(E**3 + I)
print('w=',w)
w=N(w)
print('w=',w)
```
